```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.isnull().sum()
```

```python
dataset.loc[:122, 'Region']=0
dataset.loc[122:, 'Region']=1
df=dataset
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix the spaces in columns names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

### Change the required columns as integer data types


```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float datatype

```python
# 'O' = object class
objects=[features for features in df.columns if df[features].dtype == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

### Exploratory Data Analysis

```python
## drop day, month, year
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
## Encoding of the categories in classes
df_copy['Classes'] = df_copy['Classes'].astype(str)
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
# Plot density plot for all features
df_copy.hist(bins=50, figsize=(20, 25))
plt.show()
```

```python
## Percentage for Pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting Pie chart
classlabels=['Fire', 'Not Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Pie Chart for Classes")
```

```python
## Correlation
df_copy.corr()
```

```python
import seaborn as sns
sns.heatmap(df_copy.corr())
```

```python
## Box Plot
sns.boxplot(df_copy['FWI'], color='green')
```

```python
## Monthly Fire Analysis
df['Classes'] = df['Classes'].astype(str)
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
df_temp = df.loc[df['Region']==1]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df_temp)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title("Fire Analysis of Sidi-Bel Regions", weight='bold')
```

```python
df_temp = df.loc[df['Region']==0]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df_temp)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title("Fire Analysis of Sidi-Bel Regions", weight='bold')
```
